This notebook is intended to produce the plots for the paper. The local paths are loaded from a yaml file. The input and output numpy arrays from the classifier run are expected to lie in the same folder. Create a yaml file called *plotting_directories.yml* containing the accessed method names as keys and for each a list of local directories (one for each independent run). The example below contains all the studies plotted in the notebook. However, only the one that are loaded by a cell need to be present in the yaml file. For ANODE and CATHODE, only the first classifier run will be used. For Supervised, CWoLa and Idealized AD, only one classifier training execution with multiple trainings is expected (since they only depend on the classifier and are thus already fully independent). However, in the S/B scan study where each run in on a different signal/background injection, also Idealized AD and CWoLa are independent runs. 

To create the data for the S/B scan study, use:
```
python run_data_preparation_LHCORD.py --outdir [directory to plug into the yaml file] --seed [seed] --S_over_B [S_over_B_value]
```

````python
def plot_data(savefig=None, mj1_range=(0,2), dmj_range=(0,2), nbins=50, default_color="blue", shifted_color="red"):

    bin_width_mj1 = (mj1_range[1]-mj1_range[0])/nbins
    bin_width_dmj = (dmj_range[1]-dmj_range[0])/nbins

    data = LHCORD_data_handler('innerdata_train.npy',
                               'innerdata_test.npy',
                               'outerdata_train.npy',
                               'outerdata_test.npy',
                               None)
    data.shift_data(0.1)

    full_data = np.vstack((data.original_innerdata_train, data.original_innerdata_test, data.original_outerdata_train, data.original_outerdata_test))
    full_data_shifted = np.vstack((data.innerdata_train, data.innerdata_test, data.outerdata_train, data.outerdata_test))

    fig = plt.figure(figsize=(5,6))
    plt.subplot(2, 1, 1)
    plt.hist(full_data[:,1], range=mj1_range, bins=nbins, alpha=0.7,
             label='LHCO data', zorder=1, color=default_color)
    plt.hist(full_data_shifted[:,1], range=mj1_range, bins=nbins, alpha=0.7,
             label='LHCO data, shifted', zorder=3, histtype="step", color=shifted_color)
    plt.legend(loc='upper right', frameon=False)
    plt.yscale('log')
    plt.xlabel(r'$m_{j1}$ (TeV)')
    plt.ylabel(f"Events / ({bin_width_mj1*1000:.0f} GeV)")
        
    plt.subplot(2, 1, 2)
    plt.hist(full_data[:,2], range=mj1_range, bins=nbins, alpha=0.7,
             label='LHCO data', zorder=1, color=default_color)
    plt.hist(full_data_shifted[:,2], range=mj1_range, bins=nbins, alpha=0.7,
             label='LHCO data, shifted', zorder=3, histtype="step", color=shifted_color)
    plt.legend(loc='upper right', frameon=False)
    plt.yscale('log')
    plt.xlabel(r'$\Delta m_{j}$ (TeV)')
    plt.ylabel(f"Events / ({bin_width_dmj*1000:.0f} GeV)")

    fig.tight_layout()
    if savefig is not None:
        plt.savefig(savefig, bbox_inches="tight")
        print(f"saved as {savefig}")
    plt.show()
````

````python
def plot_classifier_losses(train_losses, val_losses, train_accuracy, val_accuracy, yrange_loss=None, yrange_accuracy=None, savefig=None, suppress_show=False):
    avg_train_losses = (train_losses[5:]+train_losses[4:-1]+train_losses[3:-2]
                        +train_losses[2:-3]+train_losses[1:-4])/5
    avg_val_losses = (val_losses[5:]+val_losses[4:-1]+val_losses[3:-2]
                      +val_losses[2:-3]+val_losses[1:-4])/5
    avg_train_accuracy = (train_accuracy[5:]+train_accuracy[4:-1]+train_accuracy[3:-2]
                          +train_accuracy[2:-3]+train_accuracy[1:-4])/5
    avg_val_accuracy = (val_accuracy[5:]+val_accuracy[4:-1]+val_accuracy[3:-2]
                        +val_accuracy[2:-3]+val_accuracy[1:-4])/5
    
    fig, ax1 = plt.subplots(figsize=(7,5))
    ax1.plot(range(1, len(train_losses)), train_losses[1:], linestyle=":", color="blue")
    ax1.plot(range(1, len(val_losses)), val_losses[1:], linestyle=":", color="orange")
    ax1.plot(range(3, len(train_losses)-2), avg_train_losses, label="Training Loss", color="blue")
    ax1.plot(range(3, len(val_losses)-2), avg_val_losses, label="Validation Loss", color="orange")
    #ax1.plot(np.nan, np.nan, linestyle="None", label=" ")
    per_epoch_line, = ax1.plot(np.nan, np.nan, linestyle=":", color="black", label="Per Epoch Value")
    epoch_avg_line, = ax1.plot(np.nan, np.nan, linestyle="-", color="black", label="5-Epoch Average")

    ax2=ax1.twinx()
    ax2.plot(range(1, len(train_accuracy)), train_accuracy[1:], linestyle=":", color="purple")
    ax2.plot(range(1, len(val_accuracy)), val_accuracy[1:], linestyle=":", color="red")
    ax2.plot(range(3, len(train_accuracy)-2), avg_train_accuracy, label="Training Accuracy", linestyle="-", color="purple")
    ax2.plot(range(3, len(val_accuracy)-2), avg_val_accuracy, label="Validation Accuracy", linestyle="-", color="red")
    
    if yrange_loss is not None:
        ax1.set_ylim(*yrange_loss)
    if yrange_accuracy is not None:
        ax2.set_ylim(*yrange_accuracy)
    ax1.set_xlabel("Training Epoch")
    ax1.set_ylabel("Binary Cross Entropy Loss")
    ax1_lines, ax1_labels = ax1.get_legend_handles_labels()
    ax2_lines, ax2_labels = ax2.get_legend_handles_labels()
    ax1.legend(ax1_lines + ax2_lines, ax1_labels + ax2_labels, loc="upper center", ncol=3, frameon=False)
    ax2.set_ylabel("Accuracy")

    fig.tight_layout()
    if savefig is not None:
        plt.savefig(savefig+".pdf", bbox_inches="tight")
    if not suppress_show:
        plt.show()
    plt.close()
````

````python
def compute_median_and_variance_roc_sic(tprs_list, fprs_list, resolution=1000, mean=False):

    # interpolation
    max_min_tpr = 0.
    min_max_tpr = 1.
    for tpr in tprs_list:
        if min(tpr) > max_min_tpr:
            max_min_tpr = min(tpr)
        if max(tpr) < min_max_tpr:
            min_max_tpr = max(tpr)
    tpr_manual = np.linspace(max_min_tpr, min_max_tpr, resolution)

    roc_interpol = []
    sic_interpol = []
    for tpr, fpr in zip(tprs_list, fprs_list):
        roc_function = interp1d(tpr, 1/fpr)
        sic_function = interp1d(tpr, tpr/(fpr**(0.5)))
        roc_interpol.append(roc_function(tpr_manual))
        sic_interpol.append(sic_function(tpr_manual))
                     
    # calculating mean+std or median+quantiles
    if mean:
        roc_median = np.mean(np.stack(roc_interpol), axis=0)
        sic_median = np.mean(np.stack(sic_interpol), axis=0)
        roc_std = np.std(np.stack(roc_interpol), axis=0)
        sic_std = np.std(np.stack(sic_interpol), axis=0)
        roc_std = (roc_median-roc_std, roc_median+roc_std)
        sic_std = (sic_median-sic_std, sic_median+sic_std)
    else:
        roc_median = np.median(np.stack(roc_interpol), axis=0)
        sic_median = np.median(np.stack(sic_interpol), axis=0)
        roc_std = (np.quantile(np.stack(roc_interpol), 0.16, axis=0), np.quantile(np.stack(roc_interpol), 0.84, axis=0))
        sic_std = (np.quantile(np.stack(sic_interpol), 0.16, axis=0), np.quantile(np.stack(sic_interpol), 0.84, axis=0))
    
    return tpr_manual, roc_median, sic_median, roc_std, sic_std
````

````python
def compute_median_and_variance_roc_sic(tprs_list, fprs_list, resolution=1000, mean=False, extended_range=False, fpr_cutoff=0):

    # interpolation
    max_min_tpr = 0.
    min_max_tpr = 1.
    if not extended_range:
        for tpr in tprs_list:
            if min(tpr) > max_min_tpr:
                max_min_tpr = min(tpr)
            if max(tpr) < min_max_tpr:
                min_max_tpr = max(tpr)
    tpr_manual = np.linspace(max_min_tpr, min_max_tpr, resolution)

    roc_interpol = []
    sic_interpol = []
    for tpr, fpr in zip(tprs_list, fprs_list):
        fpr_mask = fpr>fpr_cutoff
        tpr_vals = tpr[fpr_mask]
        fpr_vals = fpr[fpr_mask]
        roc_function = interp1d(tpr_vals, 1/fpr_vals, bounds_error=False)
        sic_function = interp1d(tpr_vals, tpr_vals/(fpr_vals**(0.5)), bounds_error=False)
        roc_interpol.append(roc_function(tpr_manual))
        sic_interpol.append(sic_function(tpr_manual))
                     
    # ensure at least 5 lines for a point to be considered
    roc_matrix = np.stack(roc_interpol)
    sic_matrix = np.stack(sic_interpol)
    roc_line_count = np.count_nonzero(~np.isnan(roc_matrix), axis=0)
    sic_line_count = np.count_nonzero(~np.isnan(sic_matrix), axis=0)
    total_count = np.min(np.stack((roc_line_count, sic_line_count)), axis=0)
    roc_matrix = roc_matrix[:,total_count>=5]
    sic_matrix = sic_matrix[:,total_count>=5]
    tpr_manual = tpr_manual[total_count>=5]    
            
    # calculating mean+std or median+quantiles
    if mean:
        roc_median = np.nanmean(roc_matrix, axis=0)
        sic_median = np.nanmean(sic_matrix, axis=0)
        roc_std = np.nanstd(roc_matrix, axis=0)
        sic_std = np.nanstd(sic_matrix, axis=0)
        roc_std = (roc_median-roc_std, roc_median+roc_std)
        sic_std = (sic_median-sic_std, sic_median+sic_std)
    else:
        roc_median = np.nanmedian(roc_matrix, axis=0)
        sic_median = np.nanmedian(sic_matrix, axis=0)
        roc_std = (np.nanquantile(roc_matrix, 0.16, axis=0), np.nanquantile(roc_matrix, 0.84, axis=0))
        sic_std = (np.nanquantile(sic_matrix, 0.16, axis=0), np.nanquantile(sic_matrix, 0.84, axis=0))
    
    return tpr_manual, roc_median, sic_median, roc_std, sic_std
````

````python
def variation_plots(line_dictionary, methods=["CATHODE", "CWoLa", "ANODE", "Idealized_AD", "Supervised"], draw_bands=True, draw_single_lines=False, title_suffix=None, labels=None, color_dict=None, single_color_dict=None, sic_range=(0, 23), savefig=None, resolution=1000, suppress_random=False):
    tprs_list = []
    ROCs_list = []
    SICs_list = []
    ROC_std_list = []
    SIC_std_list = []
    colors = []
    mjj_line = False
    noise_mjj_line = False
    if color_dict is None:
        color_dict = COLOR_CODE
        single_color_dict = SINGLE_COLOR_CODE
       
    ## derive the median and the std of each method's line collection
    for method in methods:
        single_tprs = []
        single_fprs = []
        for line in line_dictionary[method]:
            single_tprs.append(line[0])
            single_fprs.append(line[1])
        tpr_manual, roc_median, sic_median, roc_std, sic_std = compute_median_and_variance_roc_sic(single_tprs, single_fprs, resolution=resolution, mean=False, extended_range=True, fpr_cutoff=1.8382353*1e-5)
        tprs_list.append(tpr_manual)
        ROCs_list.append(roc_median)
        SICs_list.append(sic_median)
        ROC_std_list.append(roc_std)
        SIC_std_list.append(sic_std)
        colors.append(color_dict[method]) 
        if "noise_mjj" in method:
            noise_mjj_line=True
        elif "mjj" in method:
            mjj_line=True
    
    if title_suffix is None:
        title_suffix = ""
    else:
        title_suffix = ", "+title_suffix
    
    if labels is None:
        labels = methods.copy()
        for i in range(len(labels)):
            labels[i] = labels[i].replace("_"," ")
    
    # draw ROCs
    plt.subplot(1,2,1)
    if draw_bands:
        for i, method in enumerate(methods):
            plt.fill_between(tprs_list[i], ROC_std_list[i][0],ROC_std_list[i][1], color=colors[i], alpha=0.3)
    if draw_single_lines:
        for i, method in enumerate(methods):
            for line in line_dictionary[method]:
                if "noise_mjj" in method:
                    linestyle = ":"
                elif "mjj" in method:
                    linestyle = "--"
                else:
                    linestyle = "-"
                plt.plot(line[0], 1/line[1], color=single_color_dict[method], linestyle=linestyle)
    for i, method in enumerate(methods):
        if "noise_mjj" in method:
            linestyle = ":"
        elif "mjj" in method:
            linestyle = "--"
        else:
            linestyle = "-"
        plt.plot(tprs_list[i], ROCs_list[i], color=colors[i], label=labels[i], linestyle=linestyle)
    if mjj_line:
        plt.plot(np.nan, np.nan, color="black", linestyle="--", label=r"using $m_{jj}$ input")
    if noise_mjj_line:
        plt.plot(np.nan, np.nan, color="black", linestyle=":", label=r"using gaussian noise input")
    elif not suppress_random:
        plt.plot(np.linspace(0.0001,1,300), 1/np.linspace(0.0001,1,300), color="black", linestyle=":", label="random")
    plt.xlabel('Signal Efficiency (True Positive Rate)')
    plt.ylabel('Rejection (1/False Positive Rate)')
    plt.title(f"Signal Region"+title_suffix, loc="right", style='italic')
    plt.legend(loc = 'upper right', frameon=False)
    plt.yscale('log')
    
    # draw SICs
    plt.subplot(1,2,2)
    if draw_bands:
        for i, method in enumerate(methods):
            plt.fill_between(tprs_list[i], SIC_std_list[i][0], SIC_std_list[i][1], color=colors[i], alpha=0.3)
    if draw_single_lines:
        for i, method in enumerate(methods):
            for line in line_dictionary[method]:
                if "noise_mjj" in method:
                    linestyle = ":"
                elif "mjj" in method:
                    linestyle = "--"
                else:
                    linestyle = "-"
                plt.plot(line[0], line[0]*(line[1]**(-0.5)), color=single_color_dict[method], linestyle=linestyle)    
    for i, method in enumerate(methods):
        if "noise_mjj" in method:
            linestyle = ":"
        elif "mjj" in method:
            linestyle = "--"
        else:
            linestyle = "-"
        plt.plot(tprs_list[i], SICs_list[i], color=colors[i], label=labels[i], linestyle=linestyle)
        print(f"max SIC of {method} is {max(SICs_list[i])}")
    if mjj_line:
        plt.plot(np.nan, np.nan, color="black", linestyle="--", label=r"using $m_{jj}$ input")
    if noise_mjj_line:
        plt.plot(np.nan, np.nan, color="black", linestyle=":", label=r"using gaussian noise input")
    elif not suppress_random:
        plt.plot(np.linspace(0.0001,1,300), np.linspace(0.0001,1,300)/np.linspace(0.0001,1,300)**(0.5), color="black", linestyle=":", label="random")
    plt.ylim(sic_range)
    plt.ylabel('Significance Improvement')
    plt.xlabel('Signal Efficiency (True Positive Rate)')
    plt.legend(loc = 'upper right', frameon=False)
    plt.title(f"Signal Region"+title_suffix, loc="right", style='italic')

    # save / display
    plt.subplots_adjust(right=2.0)
    if savefig is not None:
        plt.savefig(savefig, bbox_inches="tight")
    plt.show()
````

````python
def SIC_ratio(line_dictionary_numerator, line_dictionary_denominator, methods=["CATHODE", "CWoLa", "ANODE", "Idealized_AD", "Supervised"], draw_bands=True, title_suffix=None, labels=None, color_dict=None, ratio_range=(0, 5), savefig=None, resolution=1000):
    tprs_list = []
    ratio_list = []
    ratio_std_list = []
    colors = []
    if color_dict is None:
        color_dict = COLOR_CODE
       
    ## derive the median and the std of each method's line collection
    for method in methods:
        ### get both numerator and denominator medians and the corresponding tpr values
        single_tprs_numerator = []
        single_fprs_numerator = []
        for line in line_dictionary_numerator[method]:
            single_tprs_numerator.append(line[0])
            single_fprs_numerator.append(line[1])
        tpr_manual_numerator, _, sic_median_numerator, _, sic_std_numerator = compute_median_and_variance_roc_sic(single_tprs_numerator, single_fprs_numerator, resolution=resolution, mean=False, extended_range=True, fpr_cutoff=1.8382353*1e-5)
        single_tprs_denominator = []
        single_fprs_denominator = []
        for line in line_dictionary_denominator[method]:
            single_tprs_denominator.append(line[0])
            single_fprs_denominator.append(line[1])
        tpr_manual_denominator, _, sic_median_denominator, _, sic_std_denominator = compute_median_and_variance_roc_sic(single_tprs_denominator, single_fprs_denominator, resolution=resolution, mean=False, extended_range=True, fpr_cutoff=1.8382353*1e-5)
        
        ### interpolate onto a common tpr axis and take ratio
        min_tpr = max(min(tpr_manual_numerator), min(tpr_manual_denominator))
        max_tpr = min(max(tpr_manual_numerator), max(tpr_manual_denominator))
        tpr_manual = np.linspace(min_tpr, max_tpr, resolution)

        sic_function = interp1d(tpr_manual_numerator, sic_median_numerator)
        sic_interpol_numerator = sic_function(tpr_manual)
        sic_std0_function = interp1d(tpr_manual_numerator, sic_std_numerator[0])
        sic_std1_function = interp1d(tpr_manual_numerator, sic_std_numerator[1])
        sic_std_interpol_numerator = 0.5*(sic_std1_function(tpr_manual)-sic_std0_function(tpr_manual))
        
        sic_function = interp1d(tpr_manual_denominator, sic_median_denominator)
        sic_interpol_denominator = sic_function(tpr_manual)
        sic_std0_function = interp1d(tpr_manual_denominator, sic_std_denominator[0])
        sic_std1_function = interp1d(tpr_manual_denominator, sic_std_denominator[1])
        sic_std_interpol_denominator = 0.5*(sic_std1_function(tpr_manual)-sic_std0_function(tpr_manual))

        tprs_list.append(tpr_manual)
        ratio_list.append(sic_interpol_numerator/sic_interpol_denominator)
        ratio_std_list.append(np.sqrt((sic_std_interpol_numerator**2)/(sic_interpol_denominator**2)+(sic_std_interpol_denominator**2)*(sic_interpol_numerator**2)/(sic_interpol_denominator**4)))
        colors.append(color_dict[method]) 
      
    if title_suffix is None:
        title_suffix = ""
    else:
        title_suffix = ", "+title_suffix
    
    if labels is None:
        labels = methods.copy()
        for i in range(len(labels)):
            labels[i] = labels[i].replace("_"," ")
      
    # draw SIC ratios
    if draw_bands:
        for i, method in enumerate(methods):
            plt.fill_between(tprs_list[i], ratio_list[i]-ratio_std_list[i], ratio_list[i]+ratio_std_list[i], color=colors[i], alpha=0.3)
    for i, method in enumerate(methods):
        plt.plot(tprs_list[i], ratio_list[i], color=colors[i], label=labels[i], linestyle="-")
    plt.axhline(1., color="black", linestyle=":")
    plt.ylim(ratio_range)
    plt.ylabel('Significance Improvement Fraction')
    plt.xlabel('Signal Efficiency (True Positive Rate)')
    plt.legend(loc = 'lower right', frameon=False)
    plt.title(f"Signal Region"+title_suffix, loc="right", style='italic')

    # save / display
    #plt.subplots_adjust(right=2.0)
    if savefig is not None:
        plt.savefig(savefig, bbox_inches="tight")
    plt.show()
````

````python
def draw_correlation_matrix(data, savefig=None):
    SR_mask = np.logical_and(data[:,0]>3.3, full_data[:,0]<3.7)
    CWoLa_mask = np.logical_and(data[:,0]>3.1, full_data[:,0]<3.9)
    variable_labels = [r'$m_{jj}$', r'$m_{j1}$', r'$m_{j2}-m_{j1}$', r'$\tau_{21,1}$', r'$\tau_{21,2}$']
    vmin = -0.13
    vmax = 0.13
    
    corr_matrix_SR = pd.DataFrame(np.corrcoef(np.transpose(data[SR_mask][:,:5]))[0:1,1:], columns=variable_labels[1:5], index=variable_labels[0:1])
    corr_matrix_CWoLa = pd.DataFrame(np.corrcoef(np.transpose(data[CWoLa_mask][:,:5]))[0:1,1:], columns=variable_labels[1:5], index=variable_labels[0:1])
    corr_matrix_full = pd.DataFrame(np.corrcoef(np.transpose(data[:,:5]))[0:1,1:], columns=variable_labels[1:5], index=variable_labels[0:1])
    
    #fig=plt.figure(figsize=(12,5))
    fig = plt.figure(figsize=(6,4))
    plt.subplot(3,1,1)
    sns.heatmap(corr_matrix_SR, vmax=vmax, vmin=vmin, cmap='coolwarm', annot=True, square=True)
    fig.axes[1].set_visible(False)
    plt.title("Signal Region")

    plt.subplot(3,1,2)
    sns.heatmap(corr_matrix_CWoLa, vmax=vmax, vmin=vmin, cmap='coolwarm', annot=True, square=True)
    plt.title(r"Signal Region $\pm$ 200 GeV")
    fig.axes[3].set_visible(False)

    plt.subplot(3,1,3)
    sns.heatmap(corr_matrix_full, vmax=vmax, vmin=vmin, cmap='coolwarm', annot=True, square=True)
    plt.title(r"Full Region")
    fig.axes[5].set_visible(False)
    
    fig.tight_layout()
    if savefig is not None:
        plt.savefig(savefig, bbox_inches="tight")
        print(f"saved as {savefig}")
    plt.show()
````

````python
def max_sic_vs_S_over_B(S_B_lines, method, median=False, fpr_cutoff=0):
    max_sic_vals = {}
    for S_over_B_val in S_B_lines.keys():
        max_sic = []
        if method=="Supervised":
            lines = S_B_lines[S_over_B_val]["Supervised"][0]
            for i in range(len(lines[3][0].values())):
                tpr = list(lines[3][0].values())[i]
                fpr = list(lines[4][0].values())[i]
                fpr_mask = fpr>fpr_cutoff
                tpr_vals = tpr[fpr_mask]
                fpr_vals = fpr[fpr_mask]
                max_sic.append(max(tpr_vals*(fpr_vals**(-0.5))))
        else:
            for run in S_B_lines[S_over_B_val][method]:
                lines = run
                #for j in range(len(lines[3][0].values())):
                #    tpr = list(lines[3][0].values())[j]
                #    fpr = list(lines[4][0].values())[j]
                #    max_sic.append(max(tpr*(fpr**(-0.5))))
                tpr = list(lines[3][0].values())[0] ## only take the first classifier training
                fpr = list(lines[4][0].values())[0]
                fpr_mask = fpr>fpr_cutoff
                tpr_vals = tpr[fpr_mask]
                fpr_vals = fpr[fpr_mask]
                max_sic.append(max(tpr_vals*(fpr_vals**(-0.5))))
        if median:
            mean_max_sic = np.median(max_sic)
            std_max_sic = (np.quantile(max_sic, 0.16), np.quantile(max_sic, 0.84))
        else:
            mean_max_sic = np.mean(max_sic)
            std_max_sic = np.std(max_sic)
        max_sic_vals[S_over_B_val] = (mean_max_sic, std_max_sic)
    return max_sic_vals
````

````python
## extract mean and std of max sic

S_B_lines = {0.006: S_B_0006, 0.004: S_B_0004, 0.003: S_B_0003, 0.002: S_B_0002, 0.0015: S_B_00015, 0.001: S_B_0001, 0.0005: S_B_00005}

CATHODE_max_sics = max_sic_vs_S_over_B(S_B_lines, "CATHODE", median=True, fpr_cutoff=1.8382353*1e-5)
CWoLa_max_sics = max_sic_vs_S_over_B(S_B_lines, "CWoLa", median=True, fpr_cutoff=1.8382353*1e-5)
ANODE_max_sics = max_sic_vs_S_over_B(S_B_lines, "ANODE", median=True, fpr_cutoff=1.8382353*1e-5)
idealized_AD_max_sics = max_sic_vs_S_over_B(S_B_lines, "Idealized_AD", median=True, fpr_cutoff=1.8382353*1e-5)
supervised_max_sics = max_sic_vs_S_over_B(S_B_lines, "Supervised", median=True, fpr_cutoff=1.8382353*1e-5)
````

````python
## extract S/sqrt(B) (here taking into account only train set (approximated as half the full data))

S_B_data = {0.006: "data_S_over_B_0006", 0.004: "data_S_over_B_0004", 0.003: "data_S_over_B_0003", 0.002: "data_S_over_B_0002", 0.0015: "data_S_over_B_00015", 0.001: "data_S_over_B_0001", 0.0005: "data_S_over_B_00005"}

S_over_sqrt_B = {}
for S_over_B_val in S_B_lines.keys():
    S_over_sqrt_B_vals = []
    for path in directories[S_B_data[S_over_B_val]]:
        test_set = np.load(os.path.join(path, "innerdata_val.npy"))
        train_set = np.load(os.path.join(path, "innerdata_train.npy"))
        full_SR = np.vstack((test_set, train_set))
        S_over_sqrt_B_vals.append(sum(full_SR[:,-1]==1)/np.sqrt(sum(full_SR[:,-1]==0)))
    S_over_sqrt_B[S_over_B_val] = np.mean(S_over_sqrt_B_vals)

print(S_over_sqrt_B)
````
